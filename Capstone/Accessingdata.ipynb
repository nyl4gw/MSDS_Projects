```python
# file to access ebay.db database 
# Run these lines to get comprehensive data collection
import pandas as pd 
import sqlite3
from datetime import date
```

```python
ebaydb = sqlite3.connect("ebaydata.db")
```

```python
ebay_cursor = ebaydb.cursor()
```

```python
ebay_cursor.execute("SELECT * FROM item_information")
ebay_df = ebay_cursor.fetchall()
colnames = [x[0] for x in ebay_cursor.description]
real_df = pd.DataFrame(ebay_df, columns=colnames)
real_df #shows table in python
```

```python
ebaydb.commit()
ebaydb.close()
```

#### Run the next lines if you want to download the data as a csv:

```python
todays_date = date.today().strftime("%m/%d/%y")
```

```python
todays_date = todays_date.replace("/", "_")
```

```python
real_df.to_csv("eBaydata_updated_"+todays_date+".csv", header=True)
# name_of_table_in_python.to_csv("name_of_csv_file")
# we name the resulting csv file in accordance of the date of interest
```
