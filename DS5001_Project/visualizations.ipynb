```python
def hist_by_country(column):
    attribute = column
    plt.hist(df[df['loc'] == 'clev'][column].dropna(), alpha=0.5, label='clevland')
    plt.hist(df[df['loc'] == 'hung'][column].dropna(), alpha=0.5, label='hungary')
    plt.hist(df[df['loc'] == 'long'][column].dropna(), alpha=0.5, label='long beach')
    plt.hist(df[df['loc'] == 'swit'][column].dropna(), alpha=0.5, label='switzerland')
    plt.legend(loc='upper right')
    plt.title(column)
    plt.show()
```

```python
hist_by_country('num')
```

Comparing num (diagnosis of heart disease

- painloc - chest pain location (1 = substernal; 0 = otherwise) 
- painexer - (1 = provoked by exertion; 0 = otherwise)
- relrest - (1 = relieved after rest; 0 = otherwise)
- pncaden - sum of painloc, painexer, and relrest
- cp  - chest pain type (1 = typical angina, 2 = atypical angina, 3 = non-anginal pain, 4 = asymptomatic)

```python
df_cp = df[['num', 'cp']].copy()
df_cp['cp'].replace({1:'typical angina',2:'atypical angina', 3:'nonanginal pain', 4:'asymtomatic'}, inplace=True)
df_cp
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(x='cp', hue='num', data=df_cp)
```
